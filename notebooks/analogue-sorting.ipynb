```python
import random
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
import numpy as np
```

```python
random.seed(1)
k = 40
x_0 = random.sample(list(range(k)),k)
y_0 = (k-1)*[0.2]
z_0 = np.array(x_0 + y_0)
```

```python
def df(t,z) :
    z = list(z)
    cut = (len(z)+1)//2
    x = z[:cut]
    y = [0] + z[cut:] + [0]
    xdot = [-2*(a*a-b*b) for a,b in zip(y[:-1],y[1:])]
    ydot = [a*(b-c)for a,b,c in zip(y[1:-1],x[1:],x[:-1])]
    return(np.array(xdot + ydot))
```

```python
sol = solve_ivp(df, [0, 200], z_0,method="LSODA")
```

```python
cut = (len(z_0)+1)//2
for row in sol.y[:cut] :
    plt.plot(sol.t,row)
plt.show()
```
